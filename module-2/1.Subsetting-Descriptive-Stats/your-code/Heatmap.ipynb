## Find 3 pairs of columns, one pair that is highly positively correlated, one for which the correlation is almost zero and one pair that is negatively correlated. In each case write down an explanation why we see this correlation

1 pair of columns are highly positively correlated

```python
Can['ESTIMATED TOTAL COST'].corr(Can['NORMALIZED TOTAL COST']) 

'''
The higher the ESTIMATED TOTAL COST, the higher the NORMALIZED TOTAL COST
'''
```

1 pair of columns whose correlation is almost zero

```python
Can['MAGNITUDE'].corr(Can['FEDERAL DFAA PAYMENTS'])

'''
There is almost no correction between MAGNITUDE and FEDERAL DFAA PAYMENTS
'''
```

1 pair of columns are highly negatively correlated

```python
'''
The higher the FATALITIES, the lower the EVACUATED
'''

Can['FATALITIES'].corr(Can['EVACUATED'])

```
