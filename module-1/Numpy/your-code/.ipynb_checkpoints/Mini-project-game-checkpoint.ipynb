```python
import random
import time



def pauze():
    print('\n')
    time.sleep(2)


def intro():
    print('''
          ---------------------------------------
          ------------- Introduction ------------''')
    pauze()
    print('''
            Welcome to the game【Guess The Number】
            ''')
    pauze()
    print('''
            Please pick a number from 1-100.
            ''')
    pauze()
    print('''
            You have 8 opportunities to guess the number selected randomly by computer.
            ''')
    pauze()
    print('''
            Sounds good? Let's start!
            ''')
    pauze()



def player_number():
    while True:
        try:
            global guess
            guess=int(input('Pick a interger(1-100): '))
            break
        except ValueError:
            print('You did not give a number between 1 and 100, please give your answer again.')


def playAgain():
    while True:
        a=input('''
                  Do you want to play again? Enter 'y' to continue, enter any other buttom to exit.
                  ''')
        if a=='y':
            print('''
                    ---------------------------
                     Great, now we play again!
                    ---------------------------
                    ''')
            pauze()
            play()

        else:
            print('''
                     Thank you for palying this game, see you next time.
                     ''')
            break
    

        


def play():

    computer_number=random.randint(1,100)
    n=0
    for i in range(8):
        print('--------Round '+ str((n+1))+'----------') 
        player_number()

        n+=1
        if guess>computer_number:
            print('Too high.')
            pauze()

        elif guess<computer_number:
            print('Too low.')
            pauze()

        else:
            pauze()
            print('''
                    -----------------------------------------------------
                    Awesome! you only used {} times, congrats!   : )
                    -----------------------------------------------------
                    '''.format(n))
            break
            
            
    else:
        print('''
            =====================================
            The answer is {}.

            You are almost there.

            Because guessed wrongly for 8 times :(

            Game over.
            ======================================
            '''.format(computer_number))

    playAgain()


intro()

play()
```
